```python
import cv2
import numpy as np
import mediapipe as mp
from tensorflow.keras.models import load_model
```

```python
# Load the trained model
model = load_model('coachmodel.h5')
```

```python
# Initialize MediaPipe pose solution
mppose = mp.solutions.pose
pose = mppose.Pose()
```

```python
# Function to predict the exercise correctness
def predict_exercise(image, model):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = pose.process(image_rgb)
    if results.pose_landmarks:
        posekeypoints = np.array([[res.x, res.y, res.z, res.visibility] for res in results.pose_landmarks.landmark]).flatten()
        prediction = model.predict(np.array([posekeypoints]))[0][0]
        return prediction
    return None
```

```python
# Open the camera
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    prediction = predict_exercise(frame, model)

    if prediction is not None:
        if prediction > 0.5:
            feedback = 'Correct'
            color = (0, 255, 0)
        else:
            feedback = 'Incorrect'
            color = (0, 0, 255)

        cv2.putText(frame, feedback, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2, cv2.LINE_AA)

    cv2.imshow('Exercise Feedback', frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
